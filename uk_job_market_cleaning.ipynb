# Import

```python
import pandas as pd

df = pd.read_csv("rawData.csv")
df.head()
```

```python
print(df.shape)
print(df.columns)
df.head()
```

# Cleaning Data

```python
# clean salary column
df["salary_clean"] = (
    df["salary"]
    .astype(str)
    .str.replace("£", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.extract(r"(\d+)")
    .astype(float)
)
```

```python
# clean location column
df["location"] = (
    df["location"]
    .astype(str)
    .str.strip()
    .str.title()
)
```

```python
# drop rows missing key fields
df = df.dropna(subset=["salary_clean", "location"])
```

```python
# save cleaned file
df.to_csv("cleaned_jobs.csv", index=False)
```
